```python
import ROOT
import numpy as np
import matplotlib.pyplot as plt
```

```python

pzs=np.arange(0.0,70.0,5.0)
Err=[np.zeros_like(pzs),np.zeros_like(pzs),np.zeros_like(pzs),np.zeros_like(pzs)]
Eff=[np.zeros_like(pzs),np.zeros_like(pzs),np.zeros_like(pzs),np.zeros_like(pzs)]
Allfiles=["../muon/reco_Eta_0.250000_z500_600_eps_-6.txt.root","../muon/reco_Eta_0.330000_z500_600_eps_-6.txt.root","../muon/reco_Eta_0.430000_z500_600_eps_-6.txt.root","../muon/reco_Eta_0.530000_z500_600_eps_-6.txt.root"]
Allcolors=["red","blue","black","green"]
Alllabels=[r"$m=0.25~GeV,\epsilon=10^{-6}$",r"$m=0.33~GeV,\epsilon=10^{-6}$",r"$m=0.43~GeV,\epsilon=10^{-6}$",r"$m=0.53~GeV,\epsilon=10^{-6}$"]
fig=plt.figure(dpi=100,figsize=(10,8))
ax=fig.add_subplot(1,1,1)
for j in range(4):
    Allcounts=np.zeros_like(pzs)
    Recocounts=np.zeros_like(pzs)
    file=ROOT.TFile.Open(Allfiles[j])
    tree=file.Events
    for entry in tree:
        if entry.n_truthdimuons>0:
            #print(entry.truthdimuon_pz[0])
            ind=round(entry.truthdimuon_pz[0]/5)
            if ind>=len(pzs):
                continue
            Allcounts[ind]=Allcounts[ind]+1
            #print(entry.n_dimuons)
            if entry.n_dimuons>0:
                for k in range(entry.n_dimuons):
                    if entry.dimuon_matched[k] == 1:
                        Recocounts[ind]=Recocounts[ind]+1
                        break
    for i in range(len(pzs)):
        Err[j][i]=np.sqrt(Recocounts[i])/Allcounts[i]
        Eff[j][i]=Recocounts[i]/Allcounts[i]
        #print(Recocounts[i]/Allcounts[i])
        #print(Eff[i])
    ax.errorbar(pzs, Eff[j], yerr=Err[j], fmt='o', capsize=2, capthick=0.1,markersize=3,elinewidth=0.5, ecolor=Allcolors[j],color=Allcolors[j])
    ax.plot(pzs, Eff[j], linestyle='-', linewidth=0.5,color=Allcolors[j], label=Alllabels[j])

ax.legend()
ax.set_xlabel("$P_z$ [GeV]")
ax.set_ylabel("Reconstruction Efficiency")
#ax.set_title("Eta \u2192 A' \u2192 muons")
ax.set_title(r'$\eta \to A^{\prime} \to \mu^+\mu^-$')
fig.savefig("Eta_muon_eff_pz.pdf")
```
